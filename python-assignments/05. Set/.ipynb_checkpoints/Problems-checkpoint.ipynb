### Python Data Types: Sets - Exercises

 #### 1. Write a Python program to iteration over sets

```python
myset = {"lunes", "martes", "miercoles", "jueves", "viernes", "sabado", "domingo"}
for indice,dia in enumerate(myset): 
    print(f"{indice}:{dia}") #Remember set is unordered
```

#### 2. Write a Python program to add member(s) in a set

```python
myset = {"lunes", "martes", "miercoles", "jueves", "viernes", "sabado", "domingo"}
myset.add("fomingo")
print(myset)
```

#### 3. Write a Python program to remove item(s) from set

```python
myset.remove("fomingo")
print(myset)
```

#### 4. Write a Python program to remove an item from a set if it is present in the set.

```python
myset = {"lunes", "martes", "miercoles", "jueves", "viernes", "sabado", "domingo"}

temp = "lunes"
myset.discard(temp) #The discard() method removes the specified item from the set if the item is not present omit the error
print(myset)
```

#### 5. Write a Python program to create an intersection of sets.

```python
myset_1 = {"lunes", "martes", "miercoles", "jueves", "viernes", "sabado", "domingo"}
myset_2 = {"sabado", "domingo","abril", "mayo"}
print(myset_1 & myset_2)
```
