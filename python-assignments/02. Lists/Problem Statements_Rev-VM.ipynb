#### 2. Write a Python program to remove duplicates from a list.

```python
def rem_duplicate(lista):
    conjunto = set(lista)
    lista = list(conjunto)
    return lista


rem_duplicate([1 , 2 , 2, 1, "a", "a"])
```

#### 3. Write a Python function that takes two lists and returns True if they have at least one common member

```python
def comparar(lista1, lista2):
    flag = False
    for i in lista1:
        for j in lista2:
            if i == j:
                flag = True
    return flag
    
lista1 = [ 1, 2, 31, 41 ]
lista2 = [7, 9, 3, 4, "marac0"]

comparar(lista1, lista2)
```

#### 4. Write a Python program to get the difference between the two lists

```python
def diff(lista1, lista2):
    return list(set(lista1)-set(lista2))

lista1 = [ 1, 2, 3, 41 ]
lista2 = [7, 9, 3, 4, "casita"]
diff(lista1, lista2)
```

#### 5. Write a Python program to find the second smallest number in a list

```python
def second_small(lista):
    pos = lista.index(min(lista))
    lista.pop(pos)
    print(min(lista))
    
second_small([1, 3, 6, -3, 0, -2, -2.99])
```

#### 6.  Write a Python program to find the second largest number in a list.

```python
def second_small(lista):
    pos = lista.index(max(lista))
    lista.pop(pos)
    print(max(lista))
    
second_small([1, 3, 6, -3, 0, -2, -2.99])
```

#### 7. Write a Python program to get the frequency of the elements in a list

```python
def frequency(lista):
    counter = 0
    temp = []
    for i in range(len(lista)):
        for j in range(len(lista)):
            if lista[i] == lista[j]:
                counter += 1
        temp.append(f"{lista[i]}:{counter}") 
        counter = 0
    temp = set(temp)
    temp = list(temp)
    return temp
    
lista = [0, 1, 2, 2, 3, 4, 4, "j", "k", 0, 0, "j"]

frequency(lista)
```

#### 8. Write a Python program to convert a list of multiple integers into a single integer
* Sample list: [11, 33, 50]
* Expected Output: 113350

```python
def single_int(lista):
    cifra = ""
    for num in lista:
        cifra = cifra + str(num)
    return int(cifra)

single_int([11, 33, 50])
```

#### 9. Write a Python program to compute the similarity between two lists.
* Sample data: ["red", "orange", "green", "blue", "white"], ["black", "yellow", "green", "blue"]
* Expected Output: 
* Color1-Color2: ['white', 'orange', 'red']
* Color2-Color1: ['black', 'yellow']

```python
color1 = ["red", "orange", "green", "blue", "white"]
color2 = ["black", "yellow", "green", "blue"]
print(list(set(color1)-set(color2)))
print(list(set(color2)-set(color1)))
```

#### 10. Write a Python program to check if all dictionaries in a list are empty or not.
* Sample list : [{},{},{}]
* Return value : True
* Sample list : [{1,2},{},{}]
* Return value : False

```python
def check_dict(lista):
    for i in range(len(lista)):
        print(lista[i])
    print(all(lista))

check_dict([{},{},{}])

check_dict([{1,2},{},{}])

```
