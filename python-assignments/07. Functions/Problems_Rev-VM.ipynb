#### 2.  Write a Python function to sum all the numbers in a list

```python
def sum(lista):
    out = 0
    for i in range(len(lista)):
        out += lista[i]
    return out
lista = [10, 29, 5, 3]
print(sum(lista))
```

#### 3. Write a Python function to multiply all the numbers in a list

```python
def mul(lista):
    out = 1
    for i in range(len(lista)):
        out *= lista[i]
    return out
lista = [10, 29, 5, 3]
print(mul(lista))
```

#### 4. Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters

```python
def calc(string):
    upper = 0
    lower = 0
    for i in range(len(string)):
        if string[i].isupper():
            upper += 1
        elif string[i] == " ":
            continue
        else:
            lower += 1
    return upper, lower
string = "Gkl keie RRR"
upper, lower = calc(string)
print(f"Upper counts: {upper}")
print(f"Lowers counts: {lower}")
```

#### 7. Write a Python function that checks whether a passed string is palindrome or not

```python
def pal_det(string):
    if string == string[::-1]:
        return True
    else:
        return False

string_1 = "anilina"
string_2 = "parasito"
print(pal_det(string_1))
print(pal_det(string_2))
```
