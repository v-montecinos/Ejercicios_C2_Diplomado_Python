#### 4. Write a Python program to find the repeated items of a tuple

```python
from collections import Counter

def find_rep(tupla):
    for item in dict(Counter(tupla)).items():
        if item[1] > 1:
            print(f"el item {item[0]}, se repite {item[1]} veces")
        
tupla = 1,2,3,4,4,5,5,5,6,8,8,
find_rep(tupla)
```

#### 7. Write a Python program to replace last value of tuples in a list. 
* Sample list: [(10, 20, 40), (40, 50, 60), (70, 80, 90)]
* Expected Output: [(10, 20, 100), (40, 50, 100), (70, 80, 100)]

```python
data = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]
value = 100
out = []
for item in data:
    temp = list(item)
    temp[-1] = value  
    temp = tuple(temp)
    out.append(temp)
print(out)
```
