### Ejercicio 1: Creación de una Clase Básica y Manejo de Objetos
***
Enunciado: crea una clase llamada Empleado para modelar la información de un trabajador en una empresa. La clase debe tener un constructor que inicialice tres atributos de instancia: nombre, salario y puesto. Además, debe tener un
método mostrar_informacion() que imprima los detalles del empleado en un formato legible.

Desarrollo Sugerido:
1. Define la clase Empleado con el método constructor __init__(self,nombre, salario, puesto). 
2. Dentro del constructor, asigna los parámetros a los atributos de instancia self.nombre, self.salario y self.puesto.
3. Define un método mostrar_informacion(self) que use f-strings para imprimir el nombre, salario y puesto del empleado.
4. Crea dos objetos de la clase Empleado con información de tu elección.
5. Llama al método mostrar_informacion() en cada uno de los objetos creados.

```python
class Empleado:
    def __init__(self, nombre, salario, puesto):        
        self.nombre = nombre
        self.salario = salario
        self.puesto = puesto

    def mostrar_informacion(self):
        perfil = f"""
        Perfil del Empleado:
        -----------------------
        Nombre: {self.nombre}
        Sueldo: {self.salario:_}
        Cargo : {self.puesto}
        """
        print(perfil)

empleado_1 = Empleado("Pedro Machuca", 1_200_000, "Bodegero")
empleado_2 = Empleado("Juan Perez", 2_350_500, "Jefe de Bodega")

empleado_1.mostrar_informacion()
empleado_2.mostrar_informacion()
```

```python
# Usando @dataclass

from dataclasses import dataclass

@dataclass
class Empleado:
    nombre: str = "Mario Hugo"
    salario: int = 5_678_000
    puesto: str = "Gerente"
    
    def mostrar_informacion(self):
        perfil = f"""
Perfil del Empleado:
-----------------------
Nombre: {self.nombre}
Sueldo: {self.salario:_}
Cargo : {self.puesto}
"""
        print(perfil)

empleado_1 = Empleado(nombre = "Sr Manguera")
empleado_2 = Empleado("Juan Perez", 2_350_500, "Jefe de Bodega")

empleado_1.mostrar_informacion()
empleado_2.mostrar_informacion()
```

### Ejercicio 2: Uso de try…except…else…finally
***
Enunciado: desarrolla un programa que solicite al usuario dos números y realice una división entre ellos. El programa debe manejar de manera robusta las siguientes situaciones:
- Si el usuario ingresa un valor no numérico (ej. texto), debe capturar el error y mostrar un mensaje amigable.

- Si la operación se realiza con éxito, el programa debe imprimir el resultado dentro de un bloque else.
- Finalmente, un bloque finally debe imprimir un mensaje de finalización, sin importar el resultado de la operación.

Desarrollo Sugerido:
1. Define un bloque try.
2. Dentro del try, solicita al usuario dos entradas usando input() y conviértelas a tipo float.
3. Realiza la operación de división.
4. Define un except ValueError para manejar entradas no numéricas.
5. Define un except ZeroDivisionError para manejar la división por cero.
6. Añade un bloque else que se ejecute solo si la división es exitosa y muestre el resultado.
7. Añade un bloque finally que imprima un mensaje final.

```python
try:
    num_1 = float(input("Ingrese un mumero: "))
    num_2 = float(input("Ingrese un mumero: "))
    div = num_1 / num_2
    
except ValueError as e:
    print(f"Error debes ingresar un valor numerico -- {e}")

except ZeroDivisionError as e:
    print(f"Error division por 0 -- {e}")

else: 
    print(f"""
    {num_1}
    ------ = {num_1/num_2}
    {num_2}""")

finally:
    print("Hasta pronto")
```

### Ejercicio 3: Herencia en Clases
***
Enunciado: modifica la clase Empleado del primer ejercicio y crea una jerarquía de herencia. Crea una nueva clase llamada Gerente que herede de Empleado. La clase Gerente debe tener un atributo adicional llamado departamento y un método propio llamado asignar_tareas() que reciba una lista de tareas y las imprima.

Desarrollo Sugerido:
1. Define la clase Empleado como en el Ejercicio 1.
2. Define la clase Gerente que herede de Empleado.
3. En el constructor __init__ de Gerente, llama al constructor de la superclase Empleado usando super().__init__() para inicializar nombre y salario.
4. En el mismo constructor de Gerente, inicializa el nuevo atributo self.departamento.
5. Crea un método asignar_tareas(self, lista_tareas) en la clase Gerente que itere sobre la lista de tareas y las imprima.
6. Crea un objeto de la clase Gerente y llama tanto a un método heredado (mostrar_informacion()) como a su método propio (asignar_tareas()).

```python
class Empleado:
    def __init__(self, nombre, salario, puesto):        
        self.nombre = nombre
        self.salario = salario
        self.puesto = puesto
    
    def mostrar_informacion(self):
        perfil = f"""
Perfil del Empleado:
-----------------------
Nombre: {self.nombre}
Sueldo: {self.salario:_}
Cargo : {self.puesto}
"""
        print(perfil)
    
class Gerente(Empleado):
    def __init__(self, nombre, salario, puesto, departamento):
        super().__init__(nombre, salario, puesto)
        self.departamento = departamento

    def asignar_tareas(self, lista_tareas):
        self.lista_tareas = lista_tareas
        for tarea in lista_tareas:
            print(tarea)
            
gerente_1 = Gerente("Guaripolo", 4_560_000,"Gerente de Area", "Gerencia Tecnica")
gerente_1.mostrar_informacion()

lista_tareas = ["Presidir reunion", "Comunicar resultados la GG."]
gerente_1.asignar_tareas(lista_tareas)
```

### Ejercicio 4: Aplicación del Polimorfismo
***
Enunciado: utiliza las clases Empleado y Gerente del ejercicio anterior y crea una lista que contenga objetos de ambos tipos. Luego, define una función que itere sobre la lista y llame a un método común para cada objeto, demostrando el
polimorfismo. 

Desarrollo Sugerido:
1. Asegúrate de tener las clases Empleado y Gerente definidas.
2. Crea una lista llamada equipo que contenga una combinación de objetos Empleado y Gerente.
3. Crea una nueva clase, por ejemplo, Desarrollador, que herede de Empleado y sobreescriba el método mostrar_informacion() para que también muestre el lenguaje de programación que utiliza.
4. Añade un objeto Desarrollador a la lista equipo.
5. Crea una función llamada reporte_equipo(lista_personal) que reciba la lista como parámetro.
6. Dentro de la función, itera sobre la lista y llama al método mostrar_informacion() de cada objeto.

```python
class Empleado:
    def __init__(self, nombre, salario, puesto):        
        self.nombre = nombre
        self.salario = salario
        self.puesto = puesto
    
    def mostrar_informacion(self):
        return f"""
Perfil del Empleado:
{"=" * 33}
Nombre      : {self.nombre}
Sueldo      : ${self.salario:_}
Cargo       : {self.puesto}
"""
        
class Gerente(Empleado):
    def __init__(self, nombre, salario, puesto, departamento):
        super().__init__(nombre, salario, puesto)
        self.departamento = departamento
        
    def mostrar_informacion(self):
        return f"""
Perfil del Empleado:
{"=" * 33}
Nombre      : {self.nombre}
Sueldo      : ${self.salario:_}
Cargo       : {self.puesto}
Departamento: {self.departamento}
"""
        
class Desarrollador(Empleado):
    def __init__(self, nombre, salario, puesto, lenguaje):
        super().__init__(nombre, salario, puesto)
        self.lenguaje = lenguaje        

    def mostrar_informacion(self):
        return f"""
Perfil del Empleado:
{"=" * 33}
Nombre      : {self.nombre}
Sueldo      : ${self.salario:_}
Cargo       : {self.puesto}
lenguaje    : {self.lenguaje}
"""

empleado_1 = Empleado("Pedro Machuca", 1_200_000, "Bodegero")
empleado_2 = Empleado("Juan Perez", 2_350_500, "Jefe de Bodega")
gerente_1 = Gerente("Guaripolo", 4_560_000,"Gerente de Area", "Gerencia Tecnica")
gerente_2 = Gerente("Sr Manguera", 5_890_000, "Gerente General", "Alta Direccion")
desarrollador = Desarrollador("Mr Robot", 539_000, "Sr Developer", "Python")

equipo = [empleado_1, empleado_2, gerente_1, gerente_2, desarrollador]

def show_info(equipo):
    for i in equipo:
       print(i.mostrar_informacion())

show_info(equipo)
    
```

### Ejercicio 5: Uso de assert para Validación de Clases
***
Enunciado: modifica la clase Empleado para incluir una validación en el constructor utilizando la sentencia assert. El programa debe verificar que el salario sea un número positivo y que el nombre sea una cadena de texto no vacía. Si alguna de estas condiciones no se cumple, el programa debe lanzar un AssertionError con un mensaje claro.

Desarrollo Sugerido:
1. En el método __init__ de la clase Empleado, antes de asignar los atributos, utiliza la sentencia assert.
2. El primer assert debe verificar que isinstance(nombre, str) y que len(nombre) > 0, con un mensaje de error apropiado.
3. El segundo assert debe verificar que isinstance(salario, (int, float)) y que salario > 0, con un mensaje de error apropiado.
4. Crea un objeto Empleado con datos válidos para verificar el éxito.
5. Usa un bloque try...except para intentar crear un objeto Empleado con datos inválidos (ej. un salario negativo o un nombre vacío) y captura el AssertionError.
6. Imprime el mensaje de error capturado en el except.

```python
class Empleado:
    def __init__(self, nombre, salario, puesto):      
        assert isinstance(nombre, str) and len(nombre) > 0 , "Error! Debes ingresar un nombre valido" 
        assert isinstance(salario, (int, float)) and salario > 0 , "Error! Debes ingresar un salario valido"
        self.nombre = nombre
        self.salario = salario
        self.puesto = puesto
    
    def mostrar_info(self):
        return f"""
Perfil del Empleado:
{"=" * 33}
Nombre      : {self.nombre}
Sueldo      : ${self.salario:_}
Cargo       : {self.puesto}
"""
    
empleado_3 = Empleado("Juanin Juan Harry", 2_680_000, "Periodista")
print(empleado_3.mostrar_info())

try:
    empleado_4 = Empleado("Patana", 0, "Practicante")
    print(empleado_4.mostrar_info())
    
except AssertionError as error:
    print({error})

try:
    empleado_4 = Empleado("", 0, "Practicante")
    print(empleado_4.mostrar_info())
    
except AssertionError as error:
    print({error})

finally:
    print("Adios")
```
