### Ejercicio 1: Gestión de un inventario de productos (listas y diccionarios)
Desarrollo Sugerido:
1. Crea una lista llamada inventario que contenga tres diccionarios. Cada diccionario debe representar un producto con las siguientes claves: "nombre", "precio" y "stock". Asigna valores de tu elección.
2. Añade un cuarto producto a la lista inventario usando el método append().
3. Modifica el stock del segundo producto en la lista inventario para que sea 15
4. Calcula el valor total del inventario (precio * stock ) para cada producto y guarda los resultados en una nueva lista llamada valor_total.
5. Imprime la lista inventario final y la lista valor_total.

```python
inventario = [{ "nombre": "N macho", "precio": 3_590, "stock": 78},
               { "nombre": "SMA macho", "precio": 2_890, "stock": 124},
               { "nombre": "BNC to N", "precio": 4_680, "stock": 27}]

inventario.append({"nombre": "BNC hembra", "precio": 1_289, "stock": 200})

inventario[1].update({"stock":15})

valor_total = []
for producto in inventario:
    valor_total.append(producto["precio"] * producto["stock"])

print("Inventario:\n",inventario)
print("Valor total:\n",valor_total)
```

### Ejercicio 2: Registro de tareas pendientes (archivos de texto y with)
Desarrollo Sugerido:
1. Crea un programa que pida al usuario que ingrese tres tareas pendientes. Cada tarea debe ser una línea de texto.
2. Usa la sentencia with open(...) en modo de escritura ("w") para crear un archivo llamado "tareas.txt". Escribe cada una de las tres tareas en una línea separada.
3. Abre el mismo archivo, pero esta vez en modo de lectura ("r") con la sentencia with open(...).
4. Lee el contenido del archivo línea por línea y muestra cada tarea en la consola, asegurándote de eliminar los saltos de línea adicionales.
5. Añade un bloque try...except para manejar la posibilidad de que el archivo no exista al intentar leerlo.

```python
tareas = []
for i in range(3):
    tareas.append(str(input(f"Ingrese la tarea pendiente {i+1}:")))
try:    
    with open("tareas.txt",'w') as f:
        for tarea in tareas:
            f.write(tarea + "\n")
except IOError as e:
    print(f"Error al abrir el archivo{e}")

try:
    with open("tareas.txt",'r') as f:

        print(f.readline(),end='')
        print(f.readline(),end='')  
        print(f.readline(),end='')
        
except FileNotFoundError as e:
        print(f"Archivo no encontrado {e}")
    
except IOError as e:
    print(f"Error al abrir el archivo {e}")
    
```

### Ejercicio 3: Análisis de Datos CSV (Módulos y try...except)

Desarrollo Sugerido:
1. Crea manualmente un archivo de texto llamado "ventas.csv" con el siguiente contenido:
producto,cantidad,precio
manzana,50,1.5
banana,100,0.75
naranja,75,1.2
2. Crea un programa de Python que lea el archivo "ventas.csv" usando el módulo csv.
3. Calcula el ingreso total por cada producto (cantidad * precio) y almacena los resultados en un diccionario donde la clave sea el nombre del producto y el valor sea el ingreso.
4. Añade un bloque try...except para manejar la posibilidad de que el archivo "ventas.csv" no se encuentre (FileNotFoundError).
5. Muestra el diccionario de ingresos por producto en la consola

```python
import csv
out = {}

try:
    with open("ventas.csv", mode='r') as file:
        reader = csv.DictReader(file)
        
        for row in reader:
            nombre = row["producto"]
            ingreso = int(row["cantidad"])*float(row["precio"])
            out[nombre] = ingreso
            
        print(out)
        
except FileNotFoundError as error:
    print(f"No es posible abrir el archivo {error}")
```

### Ejercicio 4: Manejo de opciones de configuración json (json y from...import)
Desarrollo Sugerido:

1. Crea un programa que defina un diccionario de configuración. Por ejemplo:
configuracion = {
"tema": "oscuro",
"idioma": "es",
"notificaciones": True,
"usuarios_online": ["user1", "user2"]
}
2. Crea un programa que defina un diccionario de configuración. Por ejemplo:
Usa la sentencia with y la función json.dump() para guardar este diccionario en un archivo llamado "config.json". <mark> Asegúrate de usar el parámetro indent=4 para que el archivo sea legible.</mark>
3. Crea un segundo programa que use with y json.load() para leer el archivo "config.json".
4. Imprime el valor de la clave "idioma" y el primer elemento de la lista "usuarios_online".
5. Añade un try...except para manejar un posible JSONDecodeError si elarchivo no es válido.

```python
import json
from json.decoder import JSONDecodeError

config = { "tema": "oscuro", "idioma": "es", "notificaciones": True, "usuarios_online": ["user1", "user2"] }

with open("config.json", mode ='w') as file:
    json.dump(config, file, indent=4)
```

```python
try:
    with open("config.json", mode ='r') as file:
        config = json.load(file)
        
        print(json.dumps(config, sort_keys=True, indent=4))
        
        idioma = config["idioma"]
        usuario = config["usuarios_online"][0]
        
        print(f"El idioma es: {idioma}, el usuario en linea es: {usuario}")
        
except FileNotFoundError as error:
    print(f"No es posible abrir el archivo {error}")
    
except JSONDecodeError as error:
    print(f"Archivo JSON no valido {error}")        
```

### Ejercicio 5: Generador de contraseñas (módulos random y string)

Desarrollo Sugerido:
1. Crea un programa que importe los módulos random y string.
2. Define una variable caracteres que sea una combinación de letras minúsculas, mayúsculas, dígitos y signos de puntuación usando las
constantes del módulo string. Pista: puedes usar string.ascii_letters, string.digits y string.punctuation y concatenarlas.
3. Pide al usuario que ingrese la longitud deseada de la contraseña (por ejemplo, 12).
4. **Usa un bucle y random.choice()** para generar una contraseña de la longitud especificada, seleccionando caracteres al azar de la variable caracteres.
5. Imprime la contraseña generada.

```python
import random 
import string
password = ""

caracteres = string.ascii_letters + string.digits + string.punctuation
#print(caracteres)
caracteres =  ''.join(random.sample(caracteres, len(caracteres)))      #revuelve letras,digitos y signos de puntuacion.
#print(caracteres)
try:
    longitud = int(input("Ingrese la longitud deseada de la contraseña (12 por defecto): ") or 12) 
    
    for i in range(longitud):
        password += random.choice(caracteres)

    print(f"La password sugerida es: {password}")
    
except:
    print("Error la longitud debe ser un entero positivo!")
```
